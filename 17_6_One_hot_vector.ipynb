#1. Get Data

```
#input text
text=['Building some bots for Wikipedia.',
      'Wikipedia is flooded with information.',
      'There is an app for everthing.']
```

#2. Text Cleaning

```
# cleaning
import re

def clean(text):
  #lower case
  text=text.lower()

  #remove punctuations
  text=re.sub('[^a-zA-Z]'," ",text)

  return text
```

```
#call the clean function
cleaned_text=[]

for i in text:
  cleaned_text.append(clean(i))
```

```
print(cleaned_text)
```

#3. Tokenization

```
#tokenize the text
tokens=[]

for i in cleaned_text:
  tokens.append(i.split())

print(tokens)
```

#4. Vocabulary Preparation

```
#construct vocabulary
vocab=[]

for i in tokens:
  for j in i:
    if j not in vocab:
      vocab.append(j)

#remove duplicate token
vocab = list(set(vocab))

# sort tokens
vocab.sort()

print(vocab)
```

#5. Feature Representation (One Hot Encoding)

```
#one hot vectors
import numpy as np

ohe=np.zeros([len(vocab),len(vocab)])

for i in range(len(ohe)):
  ohe[i][i]=1
```
